```python
import numpy as np
from scipy.spatial.distance import cosine

def color_for_value(val, min_val=0.0, max_val=1.0):
    """
    Map val in [min_val, max_val] to an RGB color (blue to red).
    Returns an ANSI escape string for that color.
    """
    # Normalize to [0,1]
    norm = (val - min_val) / (max_val - min_val + 1e-8)
    # Interpolate RGB: blue (0,0,255) -> red (255,0,0)
    r = int(norm * 255)
    g = 0
    b = int((1 - norm) * 255)
    return f"\033[38;2;{r};{g};{b}m"

def reset_color():
    return "\033[0m"
```

```python
def show_importance(text, show_progress_bar=False):
  words = text.split(" ")
  word_embs = embed(words, show_progress_bar=show_progress_bar)
  text_emb = embed([text], show_progress_bar=show_progress_bar)[0]

  sims = [1 - cosine(w, text_emb) for w in word_embs]

  # scale min/max to spread colors
  min_sim, max_sim = min(sims), max(sims)

  for word, sim in zip(words, sims):
      color = color_for_value(sim, min_sim, max_sim)
      # print(f"{color}{word} ({sim:.3f}){reset_color()}", end=" ")
      print(f"{color}{word}{reset_color()}", end=" ")
  print()

for text in texts:
  show_importance(text)
```

```python
# for each chunk of size i..n (n from user input), embed and k means cluster.
# choose best (?) cluster
# for each input, take the longest chunk of it, if found in cluster. Remove all other chunks from this input that would have overlapped.
# repeat with the 2nd best cluster, etc
```

```python
n = 5
show_progress_bar = True
from typing import List, Tuple, Dict, Any, Tuple
import numpy as np

# 1) Get chunk embeddings ------------------------------------------------------

def get_chunk_embeddings(
    texts: List[str],
    n: int,
    embed,
    show_progress_bar: bool = True,
    pretokenize=lambda s: s.split(" "),
) -> Tuple[np.ndarray, List[Dict[str, Any]]]:
    """
    Generate all sliding-window chunks of size 1..n for each input text,
    embed them, and return (embeddings, chunks_metadata).

    Returns:
      embeddings: (num_chunks, d) array
      chunks_metadata: list of dicts with:
        {
          'text_idx': int,
          'start': int,            # start token index (inclusive)
          'end': int,              # end token index (exclusive)
          'len': int,              # number of tokens in chunk
          'chunk_text': str,       # the chunk string
        }
    """
    all_chunks: List[str] = []
    meta: List[Dict[str, Any]] = []

    for t_idx, text in enumerate(texts):
        words = pretokenize(text)
        L = len(words)
        for size in range(1, n + 1):
            if size > L:
                continue
            for start in range(0, L - size + 1):
                end = start + size
                chunk_words = words[start:end]
                chunk_text = " ".join(chunk_words)
                all_chunks.append(chunk_text)
                meta.append({
                    "text_idx": t_idx,
                    "start": start,
                    "end": end,
                    "len": size,
                    "chunk_text": chunk_text,
                })

    # Embed all chunks at once (vectorized)
    # embed() should accept List[str] and return (num_chunks, d)
    embeddings = embed(all_chunks, show_progress_bar=show_progress_bar)
    embeddings = np.asarray(embeddings)

    return embeddings, meta
```

```python
#  2) Hierarchical clustering --------------------------------------------------------

def hierarchical_clustering(
    embeddings: np.ndarray,
    similarity_cutoff: float = 0.7,
    linkage: str = 'ward',
    metric: str = 'cosine'
) -> Tuple[np.ndarray, np.ndarray, List[int], Dict[int, int]]:
    """
    Cluster embeddings with Hierarchical Clustering using similarity cutoff.

    Args:
        embeddings: (N, d) array of normalized embeddings
        similarity_cutoff: Similarity threshold for cutting the dendrogram (0-1)
                          Higher values = more clusters, lower values = fewer clusters
        linkage: Linkage criterion ('ward', 'complete', 'average', 'single')
        metric: Distance metric ('cosine', 'euclidean', etc.)

    Returns:
      labels: (N,) cluster labels
      centers: (k, d) cluster centroids (mean of each cluster)
      cluster_order: list of cluster ids from 'best' to 'worst'
                     (here: by cluster size descending)
      cluster_sizes: dict {cluster_id: size}
    """
    try:
        from sklearn.cluster import AgglomerativeClustering
        from sklearn.metrics.pairwise import cosine_distances
        from scipy.cluster.hierarchy import linkage as scipy_linkage, fcluster
        from scipy.spatial.distance import pdist
    except ImportError as e:
        raise RuntimeError(
            "scikit-learn and scipy are required for hierarchical_clustering(). "
            "Please `pip install scikit-learn scipy`."
        ) from e

    n_samples = embeddings.shape[0]
    
    if n_samples <= 1:
        # Handle edge case
        labels = np.zeros(n_samples, dtype=int)
        centers = embeddings
        cluster_order = [0] if n_samples > 0 else []
        cluster_sizes = {0: n_samples} if n_samples > 0 else {}
        return labels, centers, cluster_order, cluster_sizes

    # Compute distance matrix
    if metric == 'cosine':
        # For cosine distance, we can use the fact that embeddings are normalized
        # cosine_distance = 1 - cosine_similarity
        # Since embeddings are normalized, cosine_similarity = dot product
        distances = 1 - np.dot(embeddings, embeddings.T)
        # Ensure diagonal is 0
        np.fill_diagonal(distances, 0)
        # Convert to condensed distance matrix
        condensed_distances = pdist(distances, metric='precomputed')
    else:
        condensed_distances = pdist(embeddings, metric=metric)

    # Perform hierarchical clustering
    linkage_matrix = scipy_linkage(condensed_distances, method=linkage)
    
    # Cut the dendrogram at the similarity cutoff
    # Convert similarity cutoff to distance threshold
    # For cosine: distance = 1 - similarity, so threshold = 1 - similarity_cutoff
    if metric == 'cosine':
        distance_threshold = 1 - similarity_cutoff
    else:
        # For other metrics, we need to estimate the threshold
        # This is a rough approximation - you might want to adjust based on your data
        max_distance = np.max(condensed_distances)
        distance_threshold = max_distance * (1 - similarity_cutoff)
    
    labels = fcluster(linkage_matrix, distance_threshold, criterion='distance') - 1  # Convert to 0-based indexing
    
    # Calculate cluster centers (mean of embeddings in each cluster)
    unique_labels = np.unique(labels)
    n_clusters = len(unique_labels)
    centers = np.zeros((n_clusters, embeddings.shape[1]))
    
    for i, label in enumerate(unique_labels):
        cluster_mask = labels == label
        centers[i] = np.mean(embeddings[cluster_mask], axis=0)

    # Rank clusters by size (desc) as a simple proxy for 'best'
    cluster_sizes: Dict[int, int] = {}
    for c in unique_labels:
        cluster_sizes[c] = int((labels == c).sum())
    cluster_order = sorted(cluster_sizes.keys(), key=lambda c: cluster_sizes[c], reverse=True)

    return labels, centers, cluster_order, cluster_sizes
```

```python
# 3) Select longest non-overlapping chunks per input for a given cluster -------

def select_longest_non_overlapping_per_input(
    chunks_meta: List[Dict[str, Any]],
    labels: np.ndarray,
    cluster_id: int,
) -> Dict[int, List[Dict[str, Any]]]:
    """
    For each input text:
      - consider only chunks assigned to `cluster_id`
      - sort by length desc (ties broken by earlier start)
      - greedily select the longest chunk, removing any remaining that overlap it
      - continue until no more non-overlapping chunks remain

    Returns:
      selections: {text_idx: [chunk_meta, ...]} with non-overlapping chunks
    """
    # Group candidate chunks by input
    by_input: Dict[int, List[Dict[str, Any]]] = {}
    for m, lab in zip(chunks_meta, labels):
        if lab != cluster_id:
            continue
        by_input.setdefault(m["text_idx"], []).append(m)

    def overlap(a: Dict[str, Any], b: Dict[str, Any]) -> bool:
        # ranges [start, end) overlap iff max(start) < min(end)
        return max(a["start"], b["start"]) < min(a["end"], b["end"])

    selections: Dict[int, List[Dict[str, Any]]] = {}

    for t_idx, cand in by_input.items():
        # Sort by length desc, then by start asc
        cand = sorted(cand, key=lambda m: (-m["len"], m["start"]))
        chosen: List[Dict[str, Any]] = []
        for c in cand:
            if all(not overlap(c, sel) for sel in chosen):
                chosen.append(c)
        selections[t_idx] = chosen

    return selections
```

```python
from typing import List, Dict, Any, Tuple
import numpy as np
from colorama import Fore, Back, Style

# ---------------------------------------------------------------------------
# A) Cross-cluster non-overlapping selection (best → next → …)
#    Uses the same chunks_meta and labels from your earlier steps.
# ---------------------------------------------------------------------------

def select_non_overlapping_across_clusters(
    chunks_meta: List[Dict[str, Any]],
    labels: np.ndarray,
    cluster_order: List[int],
) -> Dict[int, List[Dict[str, Any]]]:
    """
    Implements:
      1) Choose "best" cluster (first in cluster_order).
      2) For each input text, take longest chunks from that cluster (greedy, no overlap).
      3) Remove any remaining chunks that would overlap chosen spans.
      4) Repeat with next best cluster, etc.

    Returns:
      selections_by_text: {text_idx: [ {start, end, len, chunk_text, cluster_id}, ... ] }
      Each text's selections are guaranteed non-overlapping across clusters.
    """
    # Pre-group chunks by cluster
    by_cluster: Dict[int, List[Dict[str, Any]]] = {}
    for m, lab in zip(chunks_meta, labels):
        m2 = dict(m)
        m2["cluster_id"] = int(lab)
        by_cluster.setdefault(int(lab), []).append(m2)

    # Prepare per-text final selections
    selections_by_text: Dict[int, List[Dict[str, Any]]] = {}

    def overlaps_any(a: Dict[str, Any], chosen: List[Dict[str, Any]]) -> bool:
        return any(max(a["start"], b["start"]) < min(a["end"], b["end"]) for b in chosen)

    # Work through clusters in priority order
    for cid in cluster_order:
        cand = by_cluster.get(cid, [])
        if not cand:
            continue

        # Per-text candidates for this cluster
        per_text: Dict[int, List[Dict[str, Any]]] = {}
        for m in cand:
            per_text.setdefault(m["text_idx"], []).append(m)

        # For each text, greedily add longest, non-overlapping chunks from this cluster,
        # but also respecting chunks already chosen from earlier clusters.
        for t_idx, chunks in per_text.items():
            already = selections_by_text.setdefault(t_idx, [])
            # Sort longest-first, then by earlier start
            chunks = sorted(chunks, key=lambda m: (-m["len"], m["start"]))
            for c in chunks:
                if not overlaps_any(c, already):
                    already.append(c)

    # Sort final selections per text by start (nice for rendering)
    for t_idx in selections_by_text:
        selections_by_text[t_idx].sort(key=lambda m: m["start"])

    return selections_by_text


# ---------------------------------------------------------------------------
# B) Cluster colors + rendering
# ---------------------------------------------------------------------------

def _make_color_fn(fore: str = "", back: str = ""):
    # Capture current fore/back as defaults to avoid late-binding issues
    def color_fn(x, _fore=fore, _back=back):
        return f"{_back}{_fore}{x}{Style.RESET_ALL}"
    return color_fn

def choose_cluster_colors(cluster_order: List[int]) -> Dict[int, Any]:
    """
    Assign a color function to each cluster.
    Filler words remain uncolored (i.e., plain text).
    """
    palette = [
        Fore.RED,
        Fore.BLUE,
        Fore.GREEN,
        Fore.MAGENTA,
        Fore.CYAN,
        Fore.YELLOW,
        Fore.LIGHTRED_EX,
        Fore.LIGHTBLUE_EX,
        Fore.LIGHTGREEN_EX,
        Fore.LIGHTMAGENTA_EX,
        Fore.LIGHTCYAN_EX,
        Fore.LIGHTYELLOW_EX,
    ]
    colors: Dict[int, Any] = {}
    for i, cid in enumerate(cluster_order):
        fore = palette[i % len(palette)]
        colors[cid] = _make_color_fn(fore=fore)
    return colors

def render_colored_texts(
    texts: List[str],
    selections_by_text: Dict[int, List[Dict[str, Any]]],
    cluster_colors: Dict[int, Any],
    pretokenize=lambda s: s.split(" "),
    joiner: str = " ",
) -> List[str]:
    """
    Re-create each text with colored chunks and white/unstyled filler words.
    - Chunks are colored by their cluster color.
    - Everything else is left uncolored.
    """
    outputs: List[str] = []

    for t_idx, text in enumerate(texts):
        tokens = pretokenize(text)
        L = len(tokens)

        # Mark cluster ownership per token (None = filler)
        owner: List[int] = [None] * L
        for sel in selections_by_text.get(t_idx, []):
            cid = sel["cluster_id"]
            for i in range(sel["start"], sel["end"]):
                owner[i] = cid

        # Walk tokens and wrap contiguous spans of the same owner
        out_parts: List[str] = []
        i = 0
        while i < L:
            cid = owner[i]
            j = i + 1
            while j < L and owner[j] == cid:
                j += 1

            span = joiner.join(tokens[i:j])
            if cid is None:
                out_parts.append(span)  # filler: no color
            else:
                color_fn = cluster_colors.get(cid, lambda x: x)
                out_parts.append(color_fn(span))

            i = j

        outputs.append(joiner.join(out_parts))

    return outputs
```

```python
# Import required libraries
from sklearn.manifold import TSNE
import altair as alt
import pandas as pd
import numpy as np
```

```python
# Run t-SNE on embeddings to reduce to 2D
print(f"Running t-SNE on {embs.shape[0]} embeddings with dimension {embs.shape[1]}...")

# Initialize t-SNE with reasonable parameters
tsne = TSNE(
    n_components=2,
    perplexity=min(30, embs.shape[0] - 1),  # Perplexity should be less than n_samples
    random_state=42,
    n_iter=1000,
    verbose=1
)

# Fit and transform the embeddings
tsne_results = tsne.fit_transform(embs)

print(f"t-SNE completed. Shape: {tsne_results.shape}")
```

```python
# Create DataFrame for visualization
df = pd.DataFrame({
    'x': tsne_results[:, 0],
    'y': tsne_results[:, 1],
    'text': [meta[i]['chunk_text'] for i in range(len(meta))],
    'text_idx': [meta[i]['text_idx'] for i in range(len(meta))],
    'start': [meta[i]['start'] for i in range(len(meta))],
    'end': [meta[i]['end'] for i in range(len(meta))],
    'length': [meta[i]['len'] for i in range(len(meta))],
    'cluster': labels
})

print(f"Created DataFrame with {len(df)} points")
print(f"Cluster distribution:")
print(df['cluster'].value_counts().head(10))
```

```python
# Create interactive Vega-Lite scatterplot
chart = alt.Chart(df).mark_circle(
    size=60,
    opacity=0.7
).add_selection(
    alt.selection_interval()
).encode(
    x=alt.X('x:Q', title='t-SNE Dimension 1'),
    y=alt.Y('y:Q', title='t-SNE Dimension 2'),
    color=alt.Color('cluster:N', 
                   scale=alt.Scale(scheme='category20'),
                   title='Cluster'),
    size=alt.Size('length:Q', 
                  scale=alt.Scale(range=[20, 100]),
                  title='Chunk Length'),
    tooltip=[
        alt.Tooltip('text:N', title='Text Chunk'),
        alt.Tooltip('text_idx:O', title='Text Index'),
        alt.Tooltip('start:O', title='Start Position'),
        alt.Tooltip('end:O', title='End Position'),
        alt.Tooltip('length:O', title='Length'),
        alt.Tooltip('cluster:O', title='Cluster')
    ]
).properties(
    width=800,
    height=600,
    title='t-SNE Visualization of Text Chunk Embeddings'
).interactive()

# Display the chart
chart
```
